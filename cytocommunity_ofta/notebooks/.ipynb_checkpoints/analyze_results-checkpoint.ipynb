```python
dir_tcn = "/home/owkin/project/cytocommunity_results/unsupervised/2024-10-31_14-44-14/ensemble/ImageCollection"
dir_decfrac = "/home/owkin/project/cytocommunity_results/unsupervised/chuv"
nonzero_counts_per_tcn = np.zeros(5)
enrichment_scores = np.zeros((11, 5))
for sample_id in ids_samples:
    fpath = os.path.join(dir_tcn, sample_id, "TCNLabel_MajorityVoting.csv")
    df_tcn = pd.read_csv(fpath)
    fpath = os.path.join(dir_decfrac, f"{sample_id}_DeconvolutionFrac.npy")
    decfrac_sample = np.load(fpath)

    for tcn_num in range(5):
        # +1 because python indexing starts at 0 but tcn indexing starts at 1
        ids_spots = df_tcn[df_tcn.eq(tcn_num+1)].any(axis=1).index[df_tcn.eq(tcn_num+1).any(axis=1)].tolist()
        if len(ids_spots)>0:
            df_decfrac_tcn = decfrac_sample[ids_spots,:]
            df_decfract_tcn_mean = np.mean(df_decfrac_tcn, axis=0)
            enrichment_scores[:,tcn_num] += df_decfract_tcn_mean
            nonzero_counts_per_tcn[tcn_num] += 1

# Normalize according to the number of patients for which each TCN is present
enrichment_scores = enrichment_scores / nonzero_counts_per_tcn
```

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import ttest_1samp
from statsmodels.stats.multitest import multipletests
from scipy.cluster.hierarchy import linkage, dendrogram

# Example data generation (replace this with your actual enrichment_scores)
celltypes_names = [f'CellType-{i+1}' for i in range(enrichment_scores.shape[0])]

# Create labels for x and y axes
group_labels = [f'TCN-{i+1}' for i in range(enrichment_scores.shape[1])]

# Create a DataFrame for the data
df = pd.DataFrame(enrichment_scores, columns=group_labels, index=celltypes_names)

# Compute p-values by testing each column against a baseline (e.g., 0) using a one-sample t-test
p_values = np.array([ttest_1samp(df[col], 0).pvalue for col in df.columns])

# Adjust p-values for multiple comparisons
p_values_corrected = multipletests(p_values, method='fdr_bh')[1]

# Assign significance markers based on corrected p-values
significance = np.zeros(df.shape)
significance[:, :] = np.nan  # Start with NaNs for clarity in conditions
significance[p_values_corrected < 0.1] = 0.1
significance[p_values_corrected < 0.05] = 0.05
significance[p_values_corrected < 0.01] = 0.01
significance[p_values_corrected < 0.001] = 0.001

# Perform hierarchical clustering on rows and columns
row_linkage = linkage(df, method='average')
col_linkage = linkage(df.T, method='average')

# Generate the row and column dendrograms
row_dendro = dendrogram(row_linkage, labels=df.index, no_plot=True)
col_dendro = dendrogram(col_linkage, labels=df.columns, no_plot=True)

# Reorder data based on clustering
df = df.iloc[row_dendro['leaves'], :]
df = df.iloc[:, col_dendro['leaves']]

# Reorder significance markers based on clustering
significance = significance[row_dendro['leaves'], :]
significance = significance[:, col_dendro['leaves']]

# Plotting the heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(
    df, 
    cmap="Blues", 
    linewidths=0.5, 
    cbar_kws={'label': 'Enrichment score'},
    annot=False
)

# Overlay significance markers with correct alignment
for i in range(df.shape[0]):
    for j in range(df.shape[1]):
        value = significance[i, j]
        if value == 0.1:
            plt.text(j + 0.5, i + 0.5, '*', ha='center', va='center', color='purple', fontsize=8)
        elif value == 0.05:
            plt.text(j + 0.5, i + 0.5, '**', ha='center', va='center', color='blue', fontsize=8)
        elif value == 0.01:
            plt.text(j + 0.5, i + 0.5, '***', ha='center', va='center', color='green', fontsize=8)
        elif value == 0.001:
            plt.text(j + 0.5, i + 0.5, '****', ha='center', va='center', color='red', fontsize=8)

# Customize plot
plt.xticks(rotation=90)
plt.title("Heatmap with Statistical Significance Markers")
plt.show()
```
